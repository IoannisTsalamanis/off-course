# Port distance

```r
library(lubridate)
library(plyr)
library(data.table)
library(gmapsdistance)
library(ggmap)
library(rjson)
```

```r
# set variable from config file
config_path <- sub('/4_feature_engineering','',getwd())

config <- fromJSON(file=paste(config_path, '/config.json', sep=""))

processing_path = config$DEFAULT$processing_path
port_distance_prep_filename = config$DEFAULT$port_distance_prep_filename
walking_distance_filename = config$DEFAULT$walking_distance_filename
```

```r
unique_trips <- read.csv(paste(processing_path,port_distance_prep_filename, sep=""))
unique_trips$port_name <- as.character(unique_trips$port_name)
unique_trips$last_port_name <- as.character(unique_trips$last_port_name)
```

```r
# Calculate distances between ports using walking mode. Driving mode would not work for long distances or ports in Asia
for (i in 1:length(unique_trips$last_port_name)) {
  unique_trips$port_dist[i] <- mapdist(from = unique_trips$port_name[i], to = unique_trips$last_port_name[i],
                                       mode = "walking")$km
}
```

## Export data

```r
write.csv(unique_trips, paste(processing_path,walking_distance_filename, sep=""))
```
