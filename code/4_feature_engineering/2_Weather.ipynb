# Weather Data

As strong wides, tides and visibility can affect a ships journey, weather is expected to be a significant feature in predicting delays. Ideally the weather features would be hourly however due to availabilty daily weather is used.

Download daily weather data from:
https://www7.ncdc.noaa.gov/CDO/cdoselect.cmd?datasetabbv=GSOD&countryabbv=&georegionabbv=&resolution=40
Choose United Kingdom and from the base stations choose the station closest to the port your interested in

```python
# base libraries
import pandas as pd
import os
import json
```

```python
# set variable from config file
config_path = os.path.abspath('..')

with open(config_path + '/config.json', 'r') as f:
    config = json.load(f)
    
processing_path = config['DEFAULT']['processing_path']
weather_data_download_filename = config['DEFAULT']['weather_data_download_filename']
weather_data_filename = config['DEFAULT']['weather_data_filename']
```

```python
# import data
weather = pd.read_csv(processing_path + weather_data_download_filename,header = 0,delimiter = ',')
```

```python
# transform and clean features
# tempratures are converted to centigrade
weather['TEMP'] = round((weather['TEMP'] - 32) / 1.8)
weather['DEWP'] = round((weather['DEWP'] - 32) / 1.8)
weather['MAX'] =  weather['MAX'].str.replace(' ','')
weather['MAX'] = (weather['MAX'].str.replace('*','')).astype(float)
weather['MAX'] = round((weather['MAX'] - 32) / 1.8)
weather['MIN'] =  weather['MIN'].str.replace(' ','')
weather['MIN'] = (weather['MIN'].str.replace('*','')).astype(float)
weather['MIN'] = round((weather['MIN'] - 32) / 1.8)
weather['YEARMODA'] = pd.to_datetime(weather['YEARMODA'],format='%Y%m%d')
```

## Export Data

```python
weather.to_csv(processing_path + weather_data_filename,header=True,index=False,sep=',')
```
